```python
import matplotlib.pyplot as plt  # untuk visualisasi data
from collections import Counter  # untuk menghitung modus

# Asumsi 'dataset' sudah didefinisikan sebelumnya dan berisi kolom 'Penjualan A (pcs)' dan 'Penjualan B (pcs)'
penjualan_A = dataset['Penjualan A (pcs)']
penjualan_B = dataset['Penjualan B (pcs)']

def calc_mean(penjualan_A, penjualan_B):
    # Hitung mean masing-masing variabel
    mean_A = sum(penjualan_A) / len(penjualan_A)
    mean_B = sum(penjualan_B) / len(penjualan_B)
    
    # Tampilkan hasil
    print(f"Mean Penjualan A: {mean_A}")
    print(f"Mean Penjualan B: {mean_B}")
    
    # Kembalikan hasil jika diperlukan
    return mean_A, mean_B

# Panggil fungsi calc_mean dengan dua variabel
mean_A, mean_B = calc_mean(penjualan_A, penjualan_B)

# Visualisasi Data
plt.figure(figsize=(10, 6))

# Plot distribusi Penjualan A
plt.subplot(1, 2, 1)  # 1 baris, 2 kolom, plot 1
plt.hist(penjualan_A, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribusi Penjualan A')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend([f'Mean: {mean_A:.2f}'])

# Plot distribusi Penjualan B
plt.subplot(1, 2, 2)  # 1 baris, 2 kolom, plot 2
plt.hist(penjualan_B, bins=10, color='green', alpha=0.7)
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribusi Penjualan B')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend([f'Mean: {mean_B:.2f}'])

# Tampilkan visualisasi
plt.tight_layout()
plt.show()
```

```python
# Mengonversi kolom menjadi tipe numerik
penjualan_A = pd.to_numeric(dataset['Penjualan A (pcs)'], errors='coerce')
penjualan_B = pd.to_numeric(dataset['Penjualan B (pcs)'], errors='coerce')

# Menghitung jangkauan interkuartil (IQR) untuk setiap kolom
def hitung_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return IQR

iqr_A = hitung_iqr(penjualan_A)
iqr_B = hitung_iqr(penjualan_B)

print(f'Jangkauan Interkuartil (IQR) untuk Penjualan A: {iqr_A}')
print(f'Jangkauan Interkuartil (IQR) untuk Penjualan B: {iqr_B}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat Tabel Distribusi Frekuensi
def buat_tabel_frekuensi(data, bins):
    # Menghitung frekuensi untuk setiap interval
    frekuensi, batas = np.histogram(data, bins=bins)
    tabel = pd.DataFrame({
        'Interval': [f'{batas[i]:.2f} - {batas[i+1]:.2f}' for i in range(len(batas)-1)],
        'Frekuensi': frekuensi
    })
    return tabel, batas

# Menentukan jumlah kelas (bins) untuk histogram
bins = 10

# Membuat tabel distribusi frekuensi untuk Penjualan A dan B
tabel_frekuensi_A, batas_A = buat_tabel_frekuensi(penjualan_A, bins)
tabel_frekuensi_B, batas_B = buat_tabel_frekuensi(penjualan_B, bins)

print("Tabel Distribusi Frekuensi Penjualan A:")
print(tabel_frekuensi_A)
print("\nTabel Distribusi Frekuensi Penjualan B:")
print(tabel_frekuensi_B)

# Visualisasi Histogram dan Poligon Frekuensi
plt.figure(figsize=(14, 7))

# Histogram untuk Penjualan A
plt.subplot(1, 2, 1)
plt.hist(penjualan_A, bins=bins, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogram Penjualan A')
plt.xlabel('Jumlah Penjualan (pcs)')
plt.ylabel('Frekuensi')

# Poligon Frekuensi untuk Penjualan A
titik_tengah_A = [(batas_A[i] + batas_A[i+1]) / 2 for i in range(len(batas_A)-1)]
plt.plot(titik_tengah_A, tabel_frekuensi_A['Frekuensi'], marker='o', color='blue', label='Poligon Frekuensi A')
plt.legend()

# Histogram untuk Penjualan B
plt.subplot(1, 2, 2)
plt.hist(penjualan_B, bins=bins, alpha=0.7, color='green', edgecolor='black')
plt.title('Histogram Penjualan B')
plt.xlabel('Jumlah Penjualan (pcs)')
plt.ylabel('Frekuensi')

# Poligon Frekuensi untuk Penjualan B
titik_tengah_B = [(batas_B[i] + batas_B[i+1]) / 2 for i in range(len(batas_B)-1)]
plt.plot(titik_tengah_B, tabel_frekuensi_B['Frekuensi'], marker='o', color='green', label='Poligon Frekuensi B')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import skew, mode
import numpy as np

# Contoh data (gunakan dataset yang sesuai)
penjualan_A = pd.to_numeric(dataset['Penjualan A (pcs)'], errors='coerce')
penjualan_B = pd.to_numeric(dataset['Penjualan B (pcs)'], errors='coerce')

# Menghitung mean, median, dan modus untuk Penjualan A
mean_A = penjualan_A.mean()
median_A = penjualan_A.median()
# Memastikan hasil mode diakses dengan benar
modus_A_result = mode(penjualan_A.dropna(), nan_policy='omit')
modus_A = modus_A_result.mode[0] if isinstance(modus_A_result.mode, np.ndarray) and len(modus_A_result.mode) > 0 else modus_A_result.mode

# Menghitung mean, median, dan modus untuk Penjualan B
mean_B = penjualan_B.mean()
median_B = penjualan_B.median()
modus_B_result = mode(penjualan_B.dropna(), nan_policy='omit')
modus_B = modus_B_result.mode[0] if isinstance(modus_B_result.mode, np.ndarray) and len(modus_B_result.mode) > 0 else modus_B_result.mode

# Menghitung skewness untuk Penjualan A dan B
skewness_A = skew(penjualan_A.dropna())
skewness_B = skew(penjualan_B.dropna())

# Menampilkan hasil
print(f"Penjualan A - Mean: {mean_A:.2f}, Median: {median_A:.2f}, Modus: {modus_A}")
print(f"Penjualan A - Skewness: {skewness_A:.2f}")
print(f"Penjualan B - Mean: {mean_B:.2f}, Median: {median_B:.2f}, Modus: {modus_B}")
print(f"Penjualan B - Skewness: {skewness_B:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi histogram dengan informasi skewness, mean, median, dan modus untuk Penjualan A dan B
plt.figure(figsize=(12, 6))

# Histogram untuk Penjualan A
plt.subplot(1, 2, 1)
sns.histplot(penjualan_A.dropna(), bins=30, kde=True, color='blue')
plt.axvline(mean_A, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_A:.2f}')
plt.axvline(median_A, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_A:.2f}')
if modus_A is not None:
    plt.axvline(modus_A, color='purple', linestyle='dashdot', linewidth=2, label=f'Modus: {modus_A}')
plt.title(f'Skewness Penjualan A: {skewness_A:.2f}')
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

# Histogram untuk Penjualan B
plt.subplot(1, 2, 2)
sns.histplot(penjualan_B.dropna(), bins=30, kde=True, color='green')
plt.axvline(mean_B, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_B:.2f}')
plt.axvline(median_B, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_B:.2f}')
if modus_B is not None:
    plt.axvline(modus_B, color='purple', linestyle='dashdot', linewidth=2, label=f'Modus: {modus_B}')
plt.title(f'Skewness Penjualan B: {skewness_B:.2f}')
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, kurtosis

# Membaca file Excel
dataset = pd.read_excel("Dataset.xlsx")

# Mengonversi kolom Penjualan A dan Penjualan B menjadi numerik
penjualan_A = pd.to_numeric(dataset['Penjualan A (pcs)'], errors='coerce')
penjualan_B = pd.to_numeric(dataset['Penjualan B (pcs)'], errors='coerce')

# Menghitung kurtosis untuk Penjualan A dan Penjualan B
kurtosis_A = kurtosis(penjualan_A, nan_policy='omit')
kurtosis_B = kurtosis(penjualan_B, nan_policy='omit')

# Visualisasi Penjualan A
plt.figure(figsize=(12, 6))

# Plot histogram dan fit kurva normal untuk Penjualan A
plt.subplot(1, 2, 1)
sns.histplot(penjualan_A, kde=False, color='blue', stat='density')
mu_A, std_A = penjualan_A.mean(), penjualan_A.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_A, std_A)
plt.plot(x, p, 'k', linewidth=2)
title_A = f'Penjualan A\nKurtosis: {kurtosis_A:.2f}'
plt.title(title_A)
plt.xlabel('Penjualan A (pcs)')
plt.ylabel('Density')

# Visualisasi Penjualan B
# Plot histogram dan fit kurva normal untuk Penjualan B
plt.subplot(1, 2, 2)
sns.histplot(penjualan_B, kde=False, color='green', stat='density')
mu_B, std_B = penjualan_B.mean(), penjualan_B.std()
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu_B, std_B)
plt.plot(x, p, 'k', linewidth=2)
title_B = f'Penjualan B\nKurtosis: {kurtosis_B:.2f}'
plt.title(title_B)
plt.xlabel('Penjualan B (pcs)')
plt.ylabel('Density')

# Menampilkan plot
plt.tight_layout()
plt.show()
```
